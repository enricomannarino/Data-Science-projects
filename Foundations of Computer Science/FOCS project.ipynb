### Data cleaning - *Dogs dataset*

```python
# Abbiamo notato che per le 33 osservazioni trovate non è presente una data per la variabile "posted"
# bensì altri valori che fanno riferimento alle variabili successive. 
# Infatti, a partire dalla variabile "status" fino ad "accessed" i valori risultano 'shiftati' verso sinistra.

# Controllando tutto il dataset, solo le righe in questione risultano avere questo problema. 
# Rimanendo comunque su una linea più generale, facciamo finta di non averlo notato e procediamo con la loro individuazione 
# attraverso il seguente espediente:
```

```python
errors = dogs[~dogs["posted"].str.contains("2")] # abbiamo assegnato a 'errors' le osservazioni per cui nella variabile "posted" 
# non è presente il 2 (usato come espediente perchè il 2 dovrebbe essere sempre presente come prima cifra dell'anno, 
# o comunque altrove), bensì nomi di città e, quindi, sbagliati (es. Nashville, Las vegas, ecc). 

errors.loc[:, "status":"accessed"] # visualizziamo le variabili di interesse
```

```python
var = errors.loc[:, "status":"accessed"] # selezioniamo le variabili da shiftare
```

```python
var = var.shift(periods = 1, axis = "columns") # periods=1 indica lo shifting delle colonne di una posizione
```

```python
var # ora il subset "errors" è stato corretto!
```

```python
# Vogliamo ora sostituire le righe di 'var' in quelle del dataset di partenza 'dogs'
```

```python
ind = var.index # indici utili per la sostituzione
ind
```

```python
dogs.loc[ind, "status":"accessed"] = var[:] # ora il dataset di partenza 'dogs' è corretto!

dogs.loc[ind, "status":"accessed"]
```

# 9. For each state, compute the ratio between the number of travels and the population.


```python
import us # libreria che ci permette di lavorare più facilmente con le informazioni degli stati degli USA 
```

```python
pd.set_option('display.max_rows', 100)
```

```python
travel.groupby("contact_state").count()[["id"]].reset_index()

# Notiamo che la prima osservazione per 'contact_state' fa riferimento al codice postale della Pennsylvania, pertanto provvediamo
# a sostituirlo con la sigla dello stato associato ("PA"); inoltre la sigla usata per il Nebraska è "NE" e non "NB" come 
# nel dataset.
```

```python
travel['contact_state'] = np.where(travel.contact_state == "17325", "PA", travel.contact_state) 
travel['contact_state'] = np.where(travel.contact_state == "NB", "NE", travel.contact_state)
```

```python
state_travels = travel.groupby("contact_state").count()["id"].reset_index() # corretto
state_travels = state_travels.rename(columns = {"id":"Count"})
state_travels
```

```python
diz = us.states.mapping('abbr', 'name') 
diz # grazie alla libreria US è possibile accedere al dizionario in cui, nel nostro caso, si ha come chiave la sigla e come valore
# il nome per intero dello stato (è possibile anche il viceversa).
```

```python
obs = []
for i in range(0, len(state_travels)):
        obs.append(diz[state_travels["contact_state"][i]])

obs # otteniamo il nome intero degli stati per il dataset 'travel'.
```

```python
state_travels["State"] = obs # creiamo la nuova colonna 'State' che ci permetterà di effettuare un merge col dataset 'Population'
state_travels
```

```python
new = pd.merge(left = state_travels, right = population, left_on = "State", right_on = "State")
new
```

```python
new["Population"] = new["Population"].apply(lambda x: str(x).replace(".", "")).astype(int) 
# convertiamo la variabile 'Population' in intero.
```

```python
new["Ratio"] = new["Count"] / new["Population"] # calcoliamo quindi il rapporto richiesto
new
```
